```python
from pymongo import MongoClient

cliente = MongoClient()  # Iniciando o client do mongo
banco = cliente.testes # Conectando com o banco de dados
tweets_collection = banco.tweets # Escolhendo a collection na qual os tweets serão inseridos
```

```python
dataset_list= list(tweets_collection.find())
type(dataset_list)
nova_lista=[]
tweet_dict={}
```

```python
print(dataset_list)
```

```python
for tweet in dataset_list:
    if len(tweet["pilares"]) > 1:
        tweet_dict["full_text"]=tweet["full_text"]
        tweet_dict["sentimento"]=tweet["sentimento"]
        tweet_dict["pilares"]=tweet["pilares"][(len(tweet["pilares"])-1)]
        tweet["pilares"].pop()
    nova_lista.append(tweet_dict)
    else:
        tweet_dict["full_text"]=tweet["full_text"]
        tweet_dict["sentimento"]=tweet["sentimento"]
        tweet_dict["pilares"]=tweet["pilares"][len(tweet["pilares"])-1]
    


        
```
